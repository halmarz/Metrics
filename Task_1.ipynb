```python
TP = 6
TN = 2
FN = 1
FP = 1
```

Accuracy

```python
A = (TP+TN)/(TP+TN+FP+FN)
print('Accuracy:' , A)
```

Precision

```python
P = TP / (TP+FP)
print('Precision:' , P)
```

Recall (TPR, Sensitivity)

```python
R = (TP/(TP+FN))
print('Recall:' , R)
```

F1-score

```python
F1= 2*FP / (2*FP+FP+FN)
print('F1-score:' , F1)
```

Specificity

```python
S = TN / (TN + FP)
print('Specificity:' , S)
```
